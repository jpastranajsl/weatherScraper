```python
import requests
from bs4 import BeautifulSoup

def format_city_url(city, country):
    city_formatted = city.lower().replace(' ', '-')
    country_formatted = country.lower().replace(' ', '-')
    return f"https://www.timeanddate.com/weather/{country_formatted}/{city_formatted}"

def get_weather_data(city_url):
    try:
        response = requests.get(city_url)
        response.raise_for_status()

        # Parse the HTML content with BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract current temperature
        temperature = soup.find('div', class_='h2').text.strip()

        #Extract weather description
        description = soup.find('p').text.strip()

        # # DEBUG: Print the entire section that contains weather details
        # weather_details_section = soup.find_all('p')
        # for i, p in enumerate(weather_details_section):
        #     print(f"p[{i}] text: {p.text.strip()}")  # Print index and text

        # Try using different methods to find "Feels Like"
        feels_like_tag = soup.find(string=lambda text: 'Feels Like' in text)
        #print(feels_like_tag)
        #print(f"Feels Like tag found: {feels_like_tag}")
        feels_like = feels_like_tag.split(":")[1].strip() if feels_like_tag else "Not available"
        #print(feels_like)

        # Extract forecast (inside <span> with the specific title)
        #forecast_tag = soup.find('span', title="High and low forecasted temperature today")
        forecast_tag = soup.find(string=lambda text: 'Forecast:' in text)
        #print(forecast_tag)
        #forecast = forecast_tag.text.strip() if forecast_tag else "Not available"
        forecast = forecast_tag.split(":")[1].strip() if forecast_tag else "Not available"
        #print(forecast)

        # Extract wind information
        wind_tag = soup.find(string=lambda text: 'Wind' in text)
        #print(wind_tag)
        wind_info = wind_tag.split(":")[1].strip() if wind_tag else "Not available"
        #print(wind_info)

        return {
            "Temperature": temperature,
            "Description": description,
            "Feels Like": feels_like,
            "Forecast": forecast,
            "Wind": wind_info
        }

    except requests.exceptions.RequestException as e:
        print(f"Failed to retrieve the data: {e}")
        return None
    except (AttributeError, IndexError) as e:
        print(f"Error parsing data: {e}")
        return None

def display_weather_data(weather_data):
    if weather_data:
        print(f"Current Temperature: {weather_data['Temperature']}")
        print(f"Weather Description: {weather_data['Description']}")
        print(f"Feels Like: {weather_data['Feels Like']}")
        print(f"Forecast: {weather_data['Forecast']}")
        print(f"Wind: {weather_data['Wind']}")
    else:
        print("No data available to display.")

def main():
    city = input("Enter the city: ").strip()
    country = input("Enter the country").strip()

    city_url = format_city_url(city, country)

    weather_data = get_weather_data(city_url)
    display_weather_data(weather_data)

if __name__ == "__main__":
    main()
```

```python
import requests
from bs4 import BeautifulSoup

def format_city_url(city, country):
    city_formatted = city.lower().replace(' ', '-')
    country_formatted = country.lower().replace(' ', '-')
    return f"https://www.timeanddate.com/weather/{country_formatted}/{city_formatted}"

def get_weather_data(city_url):
    try:
        response = requests.get(city_url)
        response.raise_for_status()

        # Parse the HTML content with BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract current temperature
        temperature = soup.find('div', class_='h2').text.strip()

        # Extract weather description
        description = soup.find('p').text.strip()

        # DEBUG: Print the entire section that contains weather details
        weather_details_section = soup.find_all('p')
        for i, p in enumerate(weather_details_section):
            print(f"p[{i}] text: {p.text.strip()}")  # Print index and text

        # Try using different methods to find "Feels Like"
        feels_like_tag = soup.find(string=lambda text: 'Feels Like' in text)
        if feels_like_tag:
            print(f"Feels Like tag found: {feels_like_tag}")
            feels_like = feels_like_tag.split(":")[1].strip()
        else:
            feels_like = "Not available"

        # Extract forecast (inside <span> with the specific title)
        forecast_tag = soup.find('span', title="High and low forecasted temperature today")
        forecast = forecast_tag.text.strip() if forecast_tag else "Not available"

        # Extract wind information
        wind_tag = soup.find(string=lambda text: 'Wind' in text)
        wind_info = wind_tag.split(":")[1].strip() if wind_tag else "Not available"

        return {
            "Temperature": temperature,
            "Description": description,
            "Feels Like": feels_like,
            "Forecast": forecast,
            "Wind": wind_info
        }

    except requests.exceptions.RequestException as e:
        print(f"Failed to retrieve the data: {e}")
        return None
    except (AttributeError, IndexError) as e:
        print(f"Error parsing data: {e}")
        return None

def display_weather_data(weather_data):
    if weather_data:
        print(f"Current Temperature: {weather_data['Temperature']}")
        print(f"Weather Description: {weather_data['Description']}")
        print(f"Feels Like: {weather_data['Feels Like']}")
        print(f"Forecast: {weather_data['Forecast']}")
        print(f"Wind: {weather_data['Wind']}")
    else:
        print("No data available to display.")

# input for the city and country
city = input("Enter the city: ").strip()
country = input("Enter the country: ").strip()

# URL based on user input
city_url = format_city_url(city, country)

# display weather data
weather_data = get_weather_data(city_url)
display_weather_data(weather_data)
```
